Load Dataset

```python
dataset=pd.read_csv('car_price.csv')
print(dataset)
```

```python
dataset=dataset.drop(['Serial_No'],axis=1)
dataset
```

Summarize dataset

```python
print(dataset.shape)
print(dataset.head(7))
```

Splitting dataset into X and Y
This X contain Both Numerical and Text data

```python
Xdata=dataset.drop('car_price',axis='columns')
numericalCols=Xdata.select_dtypes(exclude=['object']).columns
X=Xdata[numericalCols]
X
```

```python
Y=dataset['car_price']
Y
```

Scaling the independent variables

```python
from sklearn.preprocessing import scale 
cols=X.columns 
X.columns=cols
X
```

```python
Y
```

Splitting Dataset into Train and Test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.20,random_state=0)
```

Training using Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
# model.fit(x_train,y_train)
model.fit(x_train,y_train)
```

```python
ypred=model.predict(x_test)
from sklearn.metrics import r2_score
r2score=r2_score(y_test,ypred)
print("R2Score",r2score*100)
```
